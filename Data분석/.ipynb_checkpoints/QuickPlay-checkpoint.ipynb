# 포지션 별 영웅 분석

## 힐러 - 치유량이 가장 높은 영웅

```python
# 치유량이 가장 높은 영웅

# 두 상관관계가 높으므로 두 애트리뷰트가 높은 영웅을 선택
best_healer = df[['per_healing', 'best_healing', 'position', 'name']]
best_healer = best_healer[best_healer['position']=='healer']

best_healer.corr()
```

```python
# best_healer 정규화
def x_score(frame):
    df_std = frame.copy()
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std

z_score_heal = x_score(best_healer[['per_healing', 'best_healing']])

z_score_heal['name'] = df['name']

z_score_heal.set_index('name', inplace=True)

z_score_heal
```

```python
# 두 특성 더하기
heal_sum = z_score_heal.sum(axis=1)

heal_sum
```

## 힐러 - 피해량이 가장 높은 영웅

```python
# 피해량이 가장 높은 영웅

# 두 상관관계가 높으므로 두 애트리뷰트가 높은 영웅을 선택
best_healer = df[['per_best_dam', 'best_dam', 'position', 'name']]
best_healer = best_healer[best_healer['position']=='healer']

best_healer.corr()
```

```python
z_score_heal_deal = x_score(best_healer[['per_best_dam', 'best_dam']])

z_score_heal_deal['name'] = df['name']

z_score_heal_deal.set_index('name', inplace=True)

z_score_heal_deal
```

```python
# 두 특성 더하기
heal_sum = z_score_heal_deal.sum(axis=1)

heal_sum
```
