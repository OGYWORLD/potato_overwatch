# 포지션 별 영웅 분석

## 힐러 - 치유량이 가장 높은 영웅

```python
# 치유량이 가장 높은 영웅

# 두 상관관계가 높으므로 두 애트리뷰트가 높은 영웅을 선택
best_healer = df[['per_healing', 'best_healing', 'position', 'name']]
best_healer = best_healer[best_healer['position']=='healer']

best_healer.corr()
```

```python
# best_healer 정규화
def x_score(frame):
    df_std = frame.copy()
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std

z_score_heal = x_score(best_healer[['per_healing', 'best_healing']])

z_score_heal['name'] = df['name']

z_score_heal.set_index('name', inplace=True)

z_score_heal
```

```python
# 두 특성 더하기
heal_sum = z_score_heal.sum(axis=1)

heal_sum.sort_values(ascending=False)
```

```python
# pie chart
# pie chart를 그리기 위해 음수 값을 없애주고자 최소값의 절대값을 모든 튜플에 더해주었다.
heal_min = heal_sum.min(axis=0, skipna=True) 

heal_min = -heal_min

heal_plot = heal_sum + heal_min

name = best_healer[best_healer['position']=='healer']
labels = name['name']
ratio = heal_plot
    
#원형 그래프 
fig, ax = plt.subplots()
plt.pie(ratio,labels = labels)
plt.axis('equal')

##
plt.show()
```

```python
heal_plot
```

```python
# 1. 메르시 2. 루시우 3. 모이라
```

# 영웅추천

### 해당 포지션에 중요한 특징을 가져와서 포지션별 분석과 같은 방식으로 분석함

## 딜러

```python
# 딜러 포지션과 관련 높은 특징
#(평균 치명타, 단독처치, 연속처치, 한 게임당 최고처치, 전체 최고처치, 한 게임당 최고 준 피해량...)

most_dealer = df[['avg_critical', 'best_solo_kill', 'continue_kill', 'per_best_kill', 'best_kill', 'per_best_dam', 'best_dam', 'victory'
                 , 'position', 'name']]

most_dealer = most_dealer[most_dealer['position']=='dealer']

most_dealer
```

```python
# 튜플 값 정규화
z_score_dealer = x_score(most_dealer[['avg_critical', 'best_solo_kill', 'continue_kill', 
                                      'per_best_kill', 'best_kill', 'per_best_dam', 'best_dam', 'victory']])

# name을 인덱스로 설정
z_score_dealer['name'] = df['name']

z_score_dealer.set_index('name', inplace=True)

z_score_dealer
```

```python
# 두 특성 더하기
most_dealer = z_score_dealer.sum(axis=1)

most_dealer.sort_values(ascending=False)
```

```python
# 1.리퍼 2.시메트라 3.애쉬
```

```python
# pie chart
rec_dealer_min = most_dealer.min(axis=0, skipna=True)

rec_dealer_min = -rec_dealer_min

dealer_plot = most_dealer + rec_dealer_min

labels=df[df['position']=='dealer']
labels=labels['name']
ratio = dealer_plot
    
#원형 그래프 
fig, ax = plt.subplots()
plt.pie(ratio,labels = labels)
plt.axis('equal')

##
plt.show()
```

```python
dealer_plot
```

## 힐러

```python
# 힐러 포지션과 관련 높은 특징
#(평균 치명타, 한 게임 당 치유량, 전체 최고 치유량, 승리회수)
most_healer = df[['avg_critical', 'per_healing', 'best_healing', 'victory'
                 , 'position', 'name']]

most_healer = most_healer[most_healer['position']=='healer']

most_healer
```

```python
# 튜플 값 정규화
z_score_healer = x_score(most_healer[['avg_critical', 'per_healing', 'best_healing', 'victory']])

# name을 인덱스로
z_score_healer['name'] = df['name']

z_score_healer.set_index('name', inplace=True)

z_score_healer
```

```python
# 두 특성 더하기
most_healer = z_score_healer.sum(axis=1)

most_healer.sort_values(ascending=False)
```

```python
# pie chart
rec_healer_min = most_healer.min(axis=0, skipna=True)

rec_healer_min = -rec_healer_min

healer_plot = most_healer + rec_healer_min

labels=df[df['position']=='healer']
labels=labels['name']
ratio = healer_plot
    
#원형 그래프 
fig, ax = plt.subplots()
plt.pie(ratio,labels = labels)
plt.axis('equal')

##
plt.show()
```
