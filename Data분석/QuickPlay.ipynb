# 포지션 별 영웅 분석

## 힐러

```python
# 치유량이 가장 높은 영웅

# 두 상관관계가 높으므로 두 애트리뷰트가 높은 영웅을 선택
best_healer = df[['per_healing', 'best_healing', 'position', 'name']]
best_healer = best_healer[best_healer['position']=='healer']

best_healer.corr()
```

```python
# best_healer 정규화
def x_score(frame):
    df_std = frame.copy()
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std

x_score(best_healer[['per_healing', 'best_healing']])
```
