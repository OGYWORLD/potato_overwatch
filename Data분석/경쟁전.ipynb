# 나는 그동안 실력이 늘었는가?

```python
df_heal = df[['score', 'heal', 'win_cont', 'herodam', 'critical']]

df_heal.plot()

plt.show()
```

```python
# 정규화
def x_score(frame):
    df_std = frame.copy()
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std

df_heal = x_score(df_heal)

df_heal
```

```python
df_heal.plot()

plt.show()
```

스코어 부분이 늘긴 했지만 다른 특징에 대해서 절대적으로 비례하진 않음
늘긴 늘었으나 그것이 단순히 특징에 관해 실력이 늘었다고 보긴 힘들다.

# 내가 플래에 가려면 얼마나 해야하는가?

```python
from sklearn.linear_model import LinearRegression

X_train = df[['heal', 'win_cont', 'herodam', 'die']]

y_train = df[['score']]

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

X_test = pd.DataFrame({ # 해당 값에 따라 점수가 달라짐
    'heal' : [8000],
    'win_cont' : [0.6],
    'herodam' : [7000],
    'die' : [3],
})

predictions = lin_reg.predict(X_test)

print(predictions)
```

# 난 플래에 언제 가는가?

```python
X_train = df[['score']] # 배치점수에 따른 시즌 예측

y_train = df[['season']]

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

X_test = pd.DataFrame({
    'score' : [3200]
})

predictions = lin_reg.predict(X_test)

print(predictions)
```

```python
print(lin_reg.intercept_) # 웹페이지에 그래프 시각화 하기 위한 W와 B값 알아내기
```
