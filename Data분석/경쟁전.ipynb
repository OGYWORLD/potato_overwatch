# 나는 그동안 실력이 늘었는가?

```python
df_heal = df[['score', 'heal', 'win_cont', 'herodam', 'critical']]

df_heal.plot()

plt.show()
```

```python
# best_healer 정규화
def x_score(frame):
    df_std = frame.copy()
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std

df_heal = x_score(df_heal)

df_heal
```

```python
df_heal.plot()

plt.show()
```

# 내가 플레에 가려면 얼마나 해야하는가?

```python
from sklearn.linear_model import LinearRegression

X_train = df[['heal', 'win_cont', 'herodam', 'die']]

y_train = df[['score']]

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

X_test = pd.DataFrame({
    'heal' : [8000],
    'win_cont' : [0.6],
    'herodam' : [7000],
    'die' : [3],
})

predictions = lin_reg.predict(X_test)

print(predictions)
```

# 난 플레에 언제 가는가?

```python
X_train = df[['score']]

y_train = df[['season']]

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

X_test = pd.DataFrame({
    'score' : [3200]
})

predictions = lin_reg.predict(X_test)

print(predictions)
```
